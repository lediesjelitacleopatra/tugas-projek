```python
import numpy as np
```

```python
# Buat matrix X ukuran 2x3
X = np.array([[2, 6, 8],
              [4, 5, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
# Tampilkan ukuran matrix
print("\nUkuran matrix X:", X.shape)
```

```python
# Kolom pertama
print("\nKolom pertama:")
print(X[:, 0])
```

```python
# Baris kedua
print("\nBaris kedua:")
print(X[1, :])
```

2. ulangi soal nomor 1 menggunakan pytorch

```python
import tensorflow as tf
```

```python
# Buat matrix X ukuran 2x3
X = tf.constant([[2, 6, 8],
                 [4, 5, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
#  Tampilkan ukuran matrix
print("Shape X:", X.shape)
```

```python
# Kolom pertama
print("Kolom pertama:", X[:, 0])
```

```python
# Baris kedua
print("Baris kedua:", X[1, :])
```
